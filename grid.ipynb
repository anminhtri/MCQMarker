```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = "output_image.png"
image = cv2.imread(image_path)
plt.imshow(image)
```

```python
height, width, _ = image.shape
space = 0.11
for i in range(5):
    pt1 = (int(width*space), 0)
    pt2 = (int(width*space), height)
    space = space + 0.023
    cv2.line(image, pt1, pt2, color=(0, 255, 0), thickness=2, lineType=8, shift=0)
plt.imshow(image)
```

```python
cv2.imwrite('grid.png', image)
```
